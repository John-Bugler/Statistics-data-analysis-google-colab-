# Regresní analýza dat (text)

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt



# Regresní analýza pro každé kritérium
criteria = ['K1-lokalita', 'K2-velikost', 'K3-tvar/sklon', 'K4-stav', 'K5-IS', 'K6-ÚP', 'K7-PS/SP']

for criterion in criteria:
    X = df[criterion]
    X = sm.add_constant(X)
    y = df['JC [Kč/m2]']

    model = sm.OLS(y, X)
    results = model.fit()

    # Vykreslení grafu
    plt.figure(figsize=(8, 6))
    plt.scatter(df[criterion], df['JC [Kč/m2]'], label='Data')
    plt.plot(df[criterion], results.predict(), color='red', label='Regresní přímka')
    plt.title(f'Regresní analýza pro kritérium {criterion}')
    plt.xlabel(criterion)
    plt.ylabel('Jednotková cena [Kč/m2]')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Slovní popis vztahu mezi cenou a kritériem
    if results.pvalues[1] < 0.05:
        model_usability = "použitelný"
    else:
        model_usability = "nevhodný"

    if results.rsquared >= 0.7:
        model_strength = "velmi silnou"
    elif results.rsquared >= 0.5:
        model_strength = "silnou"
    elif results.rsquared >= 0.3:
        model_strength = "střední"
    elif results.rsquared >= 0.1:
        model_strength = "slabou"
    else:
        model_strength = "velmi slabou"

    print("\n\n")
    print(f"Regresní analýza pro kritérium {criterion}:")
    print(f"Model je {model_usability} pro popis vztahu mezi jednotkovou cenou a kritériem {criterion}.")
    print(f"Model má {model_strength} vysvětlující sílu vztahu.")
    print("\nShrnutí regresní analýzy:")
    print(results.summary())
    print("\n\n")

# Popis jednotlivých údajů v tabulce
print("Popis jednotlivých údajů v tabulce:")
print("Dependent Variable: JC [Kč/m2] - Jednotková cena v korunách za metr čtvereční.")
print("R-squared (R-kvadrát): Míra, jak dobře model vysvětluje variabilitu závislé proměnné.")
print("Adj. R-squared (Adjustovaný R-kvadrát): R-kvadrát upravený na počet prediktorů v modelu.")
print("F-statistic (F-statistika): Hodnota F-statistiky testuje celkovou statistickou významnost modelu.")
print("Prob (F-statistic) (Pravděpodobnost F-statistiky): Pravděpodobnost, že F-statistika je statisticky významná.")
print("Df Residuals (Stupně volnosti reziduí): Počet pozorování mínus počet odhadovaných parametrů modelu.")
print("Df Model (Stupně volnosti modelu): Počet odhadovaných parametrů v modelu, včetně interceptu a všech prediktorů.")
print("Covariance Type (Typ kovariance): Typ použité kovariance.")
print("Koeficienty prediktorů: Pro každý prediktor jsou uvedeny koeficienty modelu, jejich standardní chyby, hodnoty t-testu, \
    p-hodnoty a intervaly spolehlivosti.")
print("Omnibus Test (Omnibus test): Testuje normalitu reziduí.")
print("Durbin-Watson Statistic (Durbin-Watsonova statistika): Testuje přítomnost autokorelace v reziduích.")
print("Prob (Omnibus) (Pravděpodobnost Omnibus testu): Pravděpodobnost, že Omnibus test je statisticky významný.")
print("Jarque-Bera (JB) (Jarque-Bera test): Testuje normalitu reziduí pomocí testu Jarque-Bera.")
print("Skew (šikmost): Míra asymetrie reziduí.")
print("Kurtosis (špičatost): Míra, jak moc jsou rezidua koncentrovaná kolem středu.")
print("Cond. No. (Číslo podmíněnosti): Číslo podmíněnosti měří, jak moc jsou prediktory v modelu kolineární.")
```
